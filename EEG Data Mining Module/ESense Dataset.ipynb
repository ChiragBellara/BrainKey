```python
import pandas as pd
import numpy as np
```

```python
eegData = pd.read_csv('eegDataDIYA.csv')
print(eegData)
```

```python
eegDataFrame = pd.DataFrame(eegData)
eegDataFrame
```

```python
eegDataFrame = eegDataFrame.fillna(0)
eegDataFrame
```

```python
eegDataFrame.describe()
```

```python
import scipy
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
dataArray = eegDataFrame.values
print(dataArray)
```

```python
X = dataArray[:,1:11]
y = dataArray[:,10]
print(X)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1)) 
rescaledX = scaler.fit_transform(X) 
np.set_printoptions(precision=3)
print(rescaledX[0:5,:])
```

```python
scaler2 = StandardScaler() 
rescaledX2 = scaler.fit_transform(X) 
np.set_printoptions(precision=3)
print(rescaledX2[0:5,:])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Aggregating the data and creating the graphs to get an accurate value
medAlpha = eegDataFrame.loc[:,"lowAlpha":"highAlpha"]
eegDataFrame['medianAlpha'] = medAlpha.mean(axis=1)
medBeta = eegDataFrame.loc[:,"lowBeta":"highBeta"]
eegDataFrame['medianBeta'] = medBeta.mean(axis=1)
medGamma = eegDataFrame.loc[:,"lowGamma":"highGamma"]
eegDataFrame['medianGamma'] = medGamma.mean(axis=1)
eegDataFrame
```

```python
# Plot of all the factor values of the brain (alpha, beta, gamma etc.)
plt.figure(figsize = (12,6))
plt.plot(rescaledX2[1:9,:])
```

```python
# Plot of the subjects Attention & Meditation Power
plt.plot(dataArray[-1], label = "Attention")
plt.plot(dataArray[-2], label = "Meditation")
```

```python
sns.violinplot( y=eegDataFrame["attention"] )
sns.violinplot( y=eegDataFrame["meditation"] , palette="summer")
```

```python
eegEsenseData = eegDataFrame.iloc[:,1:9]
a4_dims = (12, 10)
fig, ax = plt.subplots(figsize=a4_dims)
sns.kdeplot(eegEsenseData['delta'], shade=True, bw=15, color="olive")
sns.kdeplot(eegEsenseData['theta'], shade=True, bw=15, color="blue")
sns.kdeplot(eegEsenseData['lowAlpha'], shade=True, bw=15, color="red")
sns.kdeplot(eegEsenseData['highAlpha'], shade=True, bw=15, color="yellow")
sns.kdeplot(eegEsenseData['lowBeta'], shade=True, bw=15, color="cyan")
sns.kdeplot(eegEsenseData['highBeta'], shade=True, bw=15, color="green")
sns.kdeplot(eegEsenseData['lowGamma'], shade=True, bw=15, color="tan")
sns.kdeplot(eegEsenseData['highGamma'], shade=True, bw=15, color="grey")
plt.xlim(0,2000000)
```
